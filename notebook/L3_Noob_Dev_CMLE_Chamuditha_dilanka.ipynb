````python

# Find the csv file
csv_path = None

# search all folders and find the csv file
for root,dirs, fs in os.walk('/content/ham10000'):
  for f in fs:
    if 'metadata' in f.lower() and f.endswith('.csv'):
      csv_path =os.path.join(root,f)

print(f'CSV file found : {csv_path}')

#Load it like  opening an excel file

df = pd.read_csv(csv_path)

print(f"\nTotal Photos in datase : {len(df)}")
print(f"Columns in the file      : {list(df.columns)}")
print(f"\nHow many photos  per disease:")
print(df['dx'].value_counts())
print(f"Missing Values:")
print(df.isnull().sum())
````

#Handle Missing Valeus

````python
print("\n Handle Missing Values")
print("Before : ")
print(df.isnull().sum())

median_age = df['age'].median()
# df['age'].fillna(median_age, inplace=True)
# df['sex'].fillna(df['sex'].mode()[0], inplace=True)

df['age'] = df['age'].fillna(median_age)
df['sex'] = df['sex'].fillna(df['sex'].mode()[0])

print(f"\n After : Filled missing age values with median age = {median_age}")
print(df.isnull().sum())
````

#Handle outliers

````python
print("\n Handle Outliers (age Cloumn)")

#find Quartiles
Q1 = df['age'].quantile(0.25)# lower
Q3 = df['age'].quantile(0.75)# upper
IQR = Q3 - Q1 #find spread of middle data

lower = Q1 - 1.5 *IQR
upper = Q3 + 1.5 *IQR

#count how many wrong values exist
outliear_count = len(df[(df['age'] < lower) | (df['age'] > upper)])
df['age'] = df['age'].clip(lower=lower , upper=upper)

print(f"IQR method : lower={lower},upper={upper}")
print(f"Outlier capped : {outliear_count} values")

````

#Encode Categorical Variable

````python
# encode categorical variable

disease_map = {
    'mel'  : 0,
    'nv'   : 1,
    'bcc'  : 2,
    'akiec': 3,
    'bkl'  : 4,
    'df'   : 5,
    'vasc' : 6,
}

df['dx_encoded'] = df['dx'].map(disease_map)
df['sex_encoded'] = pd.Categorical(df['sex']).codes
df['localization'] = pd.Categorical(df['localization']).codes

print(f"Disease encoding : {disease_map}")
print(f"Sex Encoded      : {df['sex_encoded'].unique()}")
print(f"Location         : {df['localization'].unique()} location -> numbers")
````

#Feature Scaling / Normalization


````python
print("\n Feature Scaling (MinMaxScaler)")
#this used normailze values
from sklearn.preprocessing import MinMaxScaler

#initialize scaling tool
scaler = MinMaxScaler()
df['age_normalized'] = scaler.fit_transform(df[['age']])
print(f"Age Before Scaling   : {df['age'].min():.1f} to {df['age'].max():.1f}")
print(f"Age after scaling    : {df['age_normalized'].min():.3f} to {df['age_normalized'].max():.3f}")
````

#Feature Enginieering

````python
# df['age'] = pd.to_numeric(df['age'],errors='coerce')
# df['age'] = df['age'].fillna(df['age'].median())

def age_group(age) :
  if age < 30 :
    return 'young'
  elif age < 50 :
    return 'middle'
  elif age < 70 :
    return 'senior'
  else :
    return 'elderly'

df['age_group'] = df['age'].apply(age_group)
df['age_group_encoded'] = pd.Categorical(df['age_group']).codes

cancer_type = ['mel','bcc','akiec']
df['is_high_risk'] = df['dx'].apply(lambda x: 1 if x in cancer_type else 0)

print(f"New feature 'age-group' :  {df['age_group'].value_counts().to_dict()}" )
print(f"New feature 'is_high_risk' : {df['is_high_risk'].value_counts().to_dict()}" )
````

#Fix Class Imbalance


````python
print("BEFORE Balancing")
print(df['dx'].value_counts())

# dx disease types cancer type

MAX_PER_CLASS = 1500

balanced_parts = []
for disease_name,group in df.groupby('dx'):
  resampled = resample(
      group,
      replace = True,
      n_samples = MAX_PER_CLASS,
      random_state = 42
  )
  balanced_parts.append(resampled)

balanced_df = pd.concat(balanced_parts).reset_index(drop=True)

print(f"\nAFTER balancing (each class = {MAX_PER_CLASS})")
print(balanced_df['dx'].value_counts())
print(f"\nTotal photos now: {len(balanced_df)}")

#Before vs after chart
fig, axes = plt.subplots(1,2,figsize=(14,5))
df['dx'].value_counts().plot(
    kind='bar',
    ax=axes[0],
    color='#E74C3C',
    title='Before Imbalanced',
    edgecolor='white'
)
balanced_df['dx'].value_counts().plot(
    kind='bar',
    ax=axes[1],
    color='#2ECC71',
    title='After Balanced',
    edgecolor='white'
)
for ax in axes:
  ax.tick_params(axis='x',rotation=45)

plt.tight_layout()
plt.show()
````

#Image Generators (Preprosseing + Argumantation)

````python
from IPython.core.display import Image
IMG_SIZE = (224,224)
BATCH_SIZE = 32


#Training normalize + random augmentation
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 20,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    zoom_range = 0.15,
    horizontal_flip = True,
    vertical_flip = True,
    fill_mode = 'nearest'
)

#Val and test :normalize only
eval_datagen = ImageDataGenerator(rescale=1./255)

def make_gen(datagen,dataframe,shuffle=True):
  return datagen.flow_from_dataframe(
      dataframe,
      x_col= 'path',
      y_col = 'dx',
      target_size = IMG_SIZE,
      batch_size = BATCH_SIZE,
      class_mode = 'categorical',
      shuffle = shuffle
      )

train_gen = make_gen(train_datagen,train_df,shuffle=True)
val_gen = make_gen(eval_datagen,val_df,shuffle=False)
test_gen = make_gen(eval_datagen,test_df,shuffle=False)

NUM_CLASSES = len(train_gen.class_indices)
print(f"Number of classes  : {NUM_CLASSES}")
print(f"Class Mapping      : {train_gen.class_indices}")

with open('class_indices.json', 'w') as f:
  json.dump(train_gen.class_indices, f, indent = 2)
print("\nclass_indices.json saved - Flask api need this file")
````

#Helper Functions

````python
def get_callbacks(name):
  return [
    EarlyStopping(
        monitor = 'val_accuracy',
        patience = 5,
        restore_best_weights = True,
        verbose =1
    ),
     ReduceLROnPlateau(
      monitor = 'val_loss',
      factor = 0.3,
      patience = 3,
      min_lr = 1e-7,
      verbose = 1
  ),
    ModelCheckpoint(
        filepath = f'{name}_best.h5',
        monitor = 'val_accuracy',
        save_best_only = True,
        verbose = 1
    ),
  ]


def plot_history(h1,h2,model_name):
    acc = h1.history['accuracy']  + h2.history['accuracy']
    val_acc = h1.history['val_accuracy']  + h2.history['val_accuracy']
    loss = h1.history['loss']  + h2.history['loss']
    val_loss = h1.history['val_loss']  + h2.history['val_loss']
    epochs = range(1,len(acc)+1)
    phase2 = len(h1.history['accuracy'])

    fig, axes = plt.subplots(1,2,figsize=(14,5))
    fig.suptitle(f"{model_name} - Training Curves", fontsize=14)

    axes[0].plot(epochs,acc,  'b-', label='Train', linewidth=2)
    axes[0].plot(epochs,val_acc, 'r--' , label='val' , linewidth=2)
    axes[0].axvline(x=phase2,color='green',linestyle=':', label='Fine-tune start')
    axes[0].set_title('Accuracy')
    axes[0].set_xlabel('Epoch')
    axes[0].legend()
    axes[0].grid(True,alpha=0.3)

    axes[1].plot(epochs,loss,  'b-', label='Train', linewidth=2)
    axes[1].plot(epochs,val_loss, 'r--' , label='val' , linewidth=2)
    axes[1].axvline(x=phase2,color='green',linestyle=':')
    axes[1].set_title('Loss')
    axes[1].set_xlabel('Epoch')
    axes[1].legend()
    axes[1].grid(True,alpha=0.3)

    plt.tight_layout()
    fname = f'curves_{model_name.lower().replace(" ", "_")}.png'
    plt.savefig(fname, dpi=150)
    plt.show()
    print(f"Best val accuracy: {max(val_acc)*100:.1f}%")

def evaluate_model(model,generator,model_name):
  loss,acc = model.evaluate(generator,verbose=0)
  print(f"\n{model_name} - Test accuracy : {acc*100:.1f}%")

  y_pred = np.argmax(model.predict(generator,verbose=1),axis = 1)
  y_true = generator.classes
  labels = list(generator.class_indices.keys())

  print("\nClassification Report")
  print(classification_report(y_true,y_pred,target_names=labels))

  cm = confusion_matrix(y_true,y_pred)
  plt.figure(figsize=(9,7))
  sns.heatmap(cm, annot= True,fmt='d',cmap='Blues',xticklabels=labels,yticklabels=labels,linewidths=0.5)
  plt.title(f"Confusion- Matrix {model_name}",fontsize=14)
  plt.ylabel('True Label')
  plt.xlabel('Predicted Label')
  plt.tight_layout()
  plt.show()
  return acc

print("Helper fucntion ready!")
````

#Train Model 1 :MobileNetV2

````python
from sklearn import metrics
print("Building MobileNetV2...")

mob_base = MobileNetV2(
    weights = 'imagenet',
    include_top = False,
    input_shape = (224,224, 3)
)

mob_base.trainable = False

x = mob_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.4)(x)
out = layers.Dense(NUM_CLASSES,activation='softmax')(x)

mob_model = models.Model(mob_base.input,out)
mob_model.compile(
    optimizer = optimizers.Adam(learning_rate=1e-3),
    #optimizer = optimizers.Adam(learning_rate=best_lr),
    loss =  'categorical_crossentropy',
    metrics = ['accuracy']
)

print(f"Total parameters : {mob_model.count_params():,}")

#Phase 1 - train head only

h1_mob = mob_model.fit(
    train_gen,
    validation_data = val_gen,
    epochs    = 5,
    callbacks = get_callbacks('mob'),
    verbose = 1
)

#Pahse 2 - fine tune last 30 layers
mob_base.trainable = True
for layer in mob_base.layers[:-30]:
  layer.trainable = False

mob_model.compile(
    optimizer = optimizers.Adam(learning_rate=1e-5),
    loss =  'categorical_crossentropy',
    metrics = ['accuracy']
)

print("\n MobileNetV2 Phase 2 fintuning")

h2_mob = mob_model.fit(
    train_gen,
    validation_data = val_gen,
    epochs    = 20,
    callbacks = get_callbacks('mob'),
    verbose = 1
)

plot_history(h1_mob,h2_mob,'MobileNetV2')
mob_model.save('mobilenet_final.h5')
print("\nMobileNetV2 saved as mobilenet_final.h5")
````
